```python
X = df.drop("target", axis=1)
```

```python
y = df["target"]
```

```python
from sklearn.preprocessing import StandardScaler
#sc_X = StandardScaler()
#x3 = sc_X.fit_transform(x3)
```

```python
from sklearn.preprocessing import LabelBinarizer
labelbin = LabelBinarizer()
```

```python
#Training Test data
from sklearn.cross_validation import train_test_split
```

```python
#spliting the data into test and train data and also putting the test size and random state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#training the model(importing the actual model)
from sklearn.linear_model import LogisticRegression
```

```python
#Instantiating the LinearRegressionModel
lm = LogisticRegression()
```

```python
X_train = X_train.drop("ID_code", axis=1)
X_test = X_test.drop("ID_code", axis=1)
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
X_train[10]
```

```python
lm.fit(X_train, y_train)
```

```python
Predictions = lm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, Predictions))
```

```python
Predictions = np.array(Predictions).reshape(-1,1)
```

```python
import csv
with open("sample_submission.csv", "w+")as my_csv:
    csvWriter = csv.writer(my_csv, delimiter=',')
    csvWriter.writerows(Predictions)
```

```python
count = 0
for i in range(len(Predictions)):
    if(Predictions[i] == y_test[i]):
        count= count + 1
percentageAccuracy = (count/len(Predictions)) * 100
print(percentageAccuracy)
```
